```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
# Definindo a variável simbólica x
x = sp.symbols('x')

# Definindo as funções f(x) e g(x)
f = x*(sp.E**(-x**2))
g = x+1

i = f - g
area = (sp.integrate(i, (x, 0, 2)))

# Converte as funções sympy para funções numpy
f = sp.lambdify(x, f, "numpy")
g = sp.lambdify(x, g, "numpy")

# Criando uma série de pontos x no intervalo de 0 a 3
x_vals = np.linspace(-1, 3, 1000)

# Calculando os valores correspondentes de f(x) e g(x)
f_vals = f(x_vals)
g_vals = g(x_vals)



# Plotando a função f(x) e g(x) no mesmo gráfico e preenchendo a área entre as curvas
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f_vals, label='$f(x) = xe^{-x^2}$')
plt.plot(x_vals, g_vals, label='$g(x) = x+1$')
plt.fill_between(x_vals, f_vals, g_vals, where=(0<=x_vals) & (x_vals<=2), color='lightblue', alpha=0.5)
plt.axvline(0, color='r', linestyle='--')  # Linha pontilhada em x = 0
plt.axvline(2, color='r', linestyle='--')  # Linha pontilhada em x = 2
plt.title('Área entre as funções $f(x)$ e $g(x)$ entre $x=0$ e $x=2$')
plt.text(0.2, 3.5, f'Área entre as curvas: ${sp.latex(area)}$')
plt.legend()
plt.show()
```

# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração.

##a)
$$\int_0^2 x^2+1 \space dx$$

##b)
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



```python
# Definindo a variável simbólica x
x = sp.symbols('x')

# Definindo as funções f(x) e g(x)
fx = (x**2)+1
gx = (6*(x**2))-(5*x)+2

af = 2
bf = 0
ag = 1
bg = -3

integral_f = (sp.integrate(fx, (x,bf,af)))
integral_g = (sp.integrate(gx, (x,bg,ag)))

f = sp.lambdify(x, fx, "numpy")
g = sp.lambdify(x, gx, "numpy")
fi = sp.lambdify(x, (sp.integrate(fx, (x))), "numpy")
gi = sp.lambdify(x, (sp.integrate(gx, (x))), "numpy")


x_vals = np.linspace(bf-1, af+1, 1000)
f_vals = f(x_vals)
fi_vals = fi(x_vals)


# Plotando a função f(x) e g(x) no mesmo gráfico e preenchendo a área entre as curvas
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f_vals, label=f'$f(x)={sp.latex(fx)}$')
plt.plot(x_vals, fi_vals, label=f'$F(x)={sp.latex(sp.integrate(fx, (x)))}$')
plt.axvline(af, color='r', linestyle='--')  # Linha pontilhada em x = 0
plt.axvline(bf, color='r', linestyle='--')  # Linha pontilhada em x = 2
plt.title('$f(x)=x^2+1$')
plt.text(0.5, 8, f'Área da curva: ${sp.latex(integral_f)}$')
plt.legend()
plt.show()

x_vals = np.linspace(bg-1, ag+1, 1000)
g_vals = g(x_vals)
gi_vals = gi(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, g_vals, label=f'$g(x)={sp.latex(gx)}$')
plt.plot(x_vals, gi_vals, label=f'$G(x)={sp.latex(sp.integrate(gx, (x)))}$')
plt.axvline(ag, color='r', linestyle='--')  # Linha pontilhada em x = 0
plt.axvline(bg, color='r', linestyle='--')  # Linha pontilhada em x = 2
plt.title('$g(x)=6x^2-5x+2$')
plt.text(-1.6, 100, f'Área da curva: ${sp.latex(integral_g)}$')
plt.legend()
plt.show()
```
